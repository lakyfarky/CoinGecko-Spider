```python
import pandas as pd
```

```python
data_cg=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/ZP/Scrapy/spider_CoinGecko/spider_CoinGecko/spiders/data_CoinGecko.csv')
data_cg
```

```python
#Izbacivanje neispravnih stupaca
data_cg.drop('Unnamed: 0',axis=1,inplace=True)
data_cg
```

```python
#Pravilno imenovanje stupaca
COLUMNS=['#','Coin','Price','1h','24h','7d','24h volume','Market capital']
data_cg.columns=COLUMNS
data_cg
```

```python
#Izbacivanje svih '\n' iz ćelija
data_cg = data_cg.replace(r'\n','',regex=True) 
data_cg
```

```python
#sortiranje po # (broj kriptovalute odnosno Market capital)
data_cg.sort_values('#',inplace=True)
data_cg
```

```python
data_cg.reset_index(drop=True,inplace=True)
data_cg
```

```python
for i in range(len(data_cg)):
  list_coin = list(str(data_cg['Coin'][i]).split(" "))
  data_cg.Coin[i]=' '.join(list_coin[:-1:])
```

```python
data_cg
```

```python
list_final=[]
for i in range(len(data_cg)):
  list_coin_names = list(str(data_cg['Coin'][i].lower()).split(" "))
  coin_name=list_coin_names
  if len(coin_name)<2:
    list_final.append(coin_name[0])
  else:
    list_final.append(str(coin_name[0]+'-'+coin_name[1]))
list_final
```

```python
import csv
with open('list_coins','w', newline='') as myfile:
     wr = csv.writer(myfile, quoting=csv.QUOTE_ALL)
     wr.writerow(list_final)
```
